### import dependencies

```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

### import csv files

```python
df1 = pd.read_csv("Data-Crawl-Twitter\data-twitter-crawler-helmi.csv")
df2 = pd.read_csv("Data-Crawl-Twitter\data2-twitter-crawler-helmi.csv")
```

```python
df1.head()
```

```python
df2.head()
```

### combine 2 dataframe

```python
df = pd.concat([df1,df2])
df
```

### data preprocessing

```python
#remove the image_url column
df = df.drop('image_url', axis=1)
df
```

```python
#lower all the text in full_text column
lowered_text = []

for text in df['full_text']:
    text = text.lower()
    lowered_text.append(text)

lowered_text
```

```python
#remove account tags from full_text column
removed_tag = []

for text in lowered_text:
    text = re.sub(r'@\w+', '', text)
    text = text.strip()
    removed_tag.append(text)

removed_tag
```

```python
#stemming the data to make the sentence to their base word
factory = StemmerFactory()
stemmer = factory.create_stemmer()

stemmed_text = []

for text in removed_tag:
    text = stemmer.stem(text)
    stemmed_text.append(text)

stemmed_text
```

### add the cleaned text to main dataframe

```python
cleaned_text = pd.DataFrame(stemmed_text, columns=["full_text_cleaned"])
cleaned_text
```

```python
df.head()
```

```python
df = df.drop('full_text', axis=1)
df['full_text_cleaned'] = cleaned_text
```

### save the dataframe to csv

```python
df.to_csv('twitter_data_preprocessed.csv')
```
